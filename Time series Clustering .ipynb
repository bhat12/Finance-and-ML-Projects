```python
def box_plot(indicators_df):
    for column in indicators_df.columns:
        sns.boxplot(indicators_df[column])
    #plt.xlim([0,10000])
        plt.title("Box plot of "+ str(column))
        plt.xlabel("Range")
        plt.show()
    
```

```python
def violin_plot(indicators_df):
    for column in indicators_df.columns:
        sns.violinplot(indicators_df[column])
        #plt.xlim([0,10000])
        plt.title("Violin plot of "+ str(column))
        plt.xlabel("Range")
        plt.show()
```

## Scaling for PCA

```python
from sklearn.preprocessing import RobustScaler
```

```python
robust=RobustScaler()
X_train=robust.fit_transform(indicators_df)
```

## K means 

```python
from sklearn.cluster import KMeans
from sklearn import metrics
results={}
```

```python
num_clusters=10
```

```python
for k in range(2,num_clusters):
    print('.'*100)
    results[k]={}
    
    # Model definition
    kmeans=KMeans(n_clusters=k).fit(X_train)
    sil_score=metrics.silhouette_score(X_train,kmeans.labels_)
    
    
    results[k]['Model']=kmeans
    results[k]['Inertia']=kmeans.inertia_
    results[k]['Silhouette']=sil_score
    results[k]['score']=kmeans.score
    
    print("Number of clusters: {}".format(k))
    print("Silhouette score: {}".format(sil_score))
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
clusters=[2,3]
for cluster in clusters:
    kmeans=KMeans(n_clusters=cluster)
    visualize=SilhouetteVisualizer(kmeans)
    visualize.fit(X_train)
    visualize.poof()
```
